```python
df.drop(["Rank", "NA_Sales", "JP_Sales", "Other_Sales"], axis=1, inplace=True)
```

```python
df
```

## <span style='color:darkred'> SÃ¼tunlarÄ± Integer'a Ã‡eviriyorum </span>

```python
df = df.astype({"Year": "str"})
```

```python
df.Year = df.Year.apply(lambda x: x.strip(".0"))
```

```python
df.Year = df.Year.apply(lambda x: x.replace("201", "2010"))
```
