```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
df = pd.read_csv('Oyun.csv')
```

```python
df
```

```python
df.info()
```


---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------
## <span style='color:darkred'> Kullanmayacağım Sütunları Atıyorum </span>
---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------

```python
df.drop(["Rank", "NA_Sales", "JP_Sales", "Other_Sales"], axis=1, inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(["index"], axis=1, inplace=True)
```

```python
df
```


---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------
## <span style='color:darkred'> Sütunları Integer'a Çeviriyorum </span>
---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------

```python
df = df.astype({"Year": "str"})
```

```python
df.Year = df.Year.apply(lambda x: x.strip(".0"))
```

```python
df.Year = df.Year.apply(lambda x: x.replace("201", "2010"))
```

```python
df
```


---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------
## <span style='color:darkred'> Tür Alanında Global Satışları Topluyorum </span>
---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------

```python
g_sum = df.groupby(["Genre"])[["Global_Sales"]].sum()
```

```python
e_sum
```

```python
g_ort
```

```python
plt.figure(figsize=[8,6])
plt.pie(g_sum.Global_Sales, labels=["Action", "Adventure", "Fighting", "Misc", "Platform", "Puzzle", "Racing", "Role-Playing", "Shooter", "Simulation", "Sports", "Strategy"], autopct="%.2f", shadow=True, radius=1);
```


---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------
## <span style='color:darkred'> Tür Alanında Avrupa Satışlarını Topluyorum </span>
---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------

```python
e_sum = df.groupby(["Genre"])[["EU_Sales"]].sum()
```

```python
plt.figure(figsize=[8,6])
plt.pie(e_sum.EU_Sales, labels=["Action", "Adventure", "Fighting", "Misc", "Platform", "Puzzle", "Racing", "Role-Playing", "Shooter", "Simulation", "Sports", "Strategy"], autopct="%.2f", shadow=True, radius=1);
```


---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------
## <span style='color:darkred'> Global'deki Toplam Satış Grafiği </span>
---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=[8,6])
sns.barplot(x=g_sum.index, y=g_sum.Global_Sales, data=df, palette="rocket", order=g_sum.sort_values("Global_Sales", ascending=False).index)
plt.xticks(rotation=45);
plt.xlabel("Genre", fontsize=15, weight="bold", color="darkblue")
plt.ylabel("Sales", fontsize=15, weight="bold", color="darkred")
plt.title("Global'de Toplam En Çok Satılan Oyun Türleri 2022", fontsize=20, weight="bold", color="darkorange");
```


---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------
## <span style='color:darkred'> Avrupa'daki Toplam Satış Grafiği </span>
---------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=[8,6])
sns.barplot(x=e_sum.index, y=e_sum.EU_Sales, data=df, palette="rocket", order=e_sum.sort_values("EU_Sales", ascending=False).index)
plt.xticks(rotation=45);
plt.xlabel("Genre", fontsize=15, weight="bold", color="darkblue")
plt.ylabel("Sales", fontsize=15, weight="bold", color="darkred")
plt.title("Avrupa'da Toplam En Çok Satılan Oyun Türleri 2022", fontsize=20, weight="bold", color="darkorange");
```
